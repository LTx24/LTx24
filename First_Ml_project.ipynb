# Loading Data

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

# **Data preparation**

### Data seperation as X and Y


```python
y=df['logS']
y
```

```python
#separting inputs as X by removing logs column.

x=df.drop('logS',axis=1)  # axis=1 will allow drop function to work with data as column mode,
                   # however if we use axis equal to 0,it will work it in the row mode.
x
```

### Data Splitting


```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_test
```

# **Model Building**

### Linear Regression

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
```

### **Applying the model to make a prediction**

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
print(y_lr_train_pred)
```

```python
y_lr_test_pred
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error,r2_score

lr_train_mse=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_mse=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
print("LR MSE (TRAIN):",lr_train_mse)
print("LR R2 (TRAIN):",lr_train_r2)
print("LR MSE (TEST):",lr_test_mse)
print("LR R2 (TEST):",lr_test_r2)
```

```python
lr_results=pd.DataFrame(['Linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
lr_results
```

## **Random Forest**

### Training the Model

```python
#since, our output (y) is a quantitative variable, we are going to build a regression model.

from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

### Applying the model to make a prediction

```python
y_rf_train_pred=rf.predict(x_train)
y_rf_test_pred=rf.predict(x_test)
```

### Evaluate the Model Performance

```python
from sklearn.metrics import mean_squared_error,r2_score

rf_train_mse=mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2=r2_score(y_train,y_rf_train_pred)

rf_test_mse=mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2=r2_score(y_test,y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```

## **Model Comparison**

```python
df_models=pd.concat([lr_results,rf_results],axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```

# **Data Visualisation of Predicted Results**

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train,y=y_lr_train_pred,c="#7CAE00",alpha=0.3)

z=np.polyfit(y_train,y_lr_train_pred,1)
p=np.poly1d(z)

plt.plot(y_train,p(y_train),"#F8766D")
plt.ylabel("Predicted logS")
plt.xlabel("Real logS")
plt.show()
```
